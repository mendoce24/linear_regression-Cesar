Get data from csv file.

```python
import pandas as pd
import numpy as np

immo = pd.read_csv("../data/_data_clean.csv", index_col='id')
display(immo.head())
display(immo.isnull().mean() * 100)#percentage of missing values per column
```

```python
immo.select_dtypes(include=np.number)
```

```python
immo = immo[['price', 'floor', 'bedroomCount', 'netHabitableSurface', 
            'constructionYear', 'facadeCount', 'floorCount', 'gardenSurface', 'terraceSurface', 'land', 'bathroomCount', 
            'showerRoomCount', 'toiletCount', 'parkingCountIndoor', 'parkingCountOutdoor', 'parkingCountClosedBox', 
            'primaryEnergyConsumptionPerSqm', 'price_per_sm']]

immo = immo[~immo['constructionYear'].isna()]

immo.shape
immo.head()
```

```python
display(immo.isnull().mean() * 100)#percentage of missing values per column
```
